```python
from google.colab import drive
drive.mount('/content/drive')
import numpy as np
import matplotlib.pyplot as plt
import skimage as ski
import rasterio
import rasterio.plot
#https://developers.google.com/earth-engine/apidocs/ee-image-getdownloadurl#colab-python
#https://help.marine.copernicus.eu/en/articles/5051975-how-to-open-and-visualize-geotiff-files-using-pythonhttps://help.marine.copernicus.eu/en/articles/5051975-how-to-open-and-visualize-geotiff-files-using-python

#https://datacarpentry.org/image-processing/07-thresholding.html
#https://help.marine.copernicus.eu/en/articles/5051975-how-to-open-and-visualize-geotiff-files-using-python
#https://saturncloud.io/blog/python-a-comprehensive-guide-to-numpyhistogram-plot/
```

```python
tiff = rasterio.open("/content/drive/MyDrive/gee/SAR_1.tif")
rasterio.plot.show(tiff, title = "SAR image visualization")
```

```python
image_array = tiff.read(1)
```

```python
# show the histogram of the blurred image
histogram, bin_edges = np.histogram(image_array, bins=256, range=(-40.0, 0))
fig, ax = plt.subplots()
plt.plot(bin_edges[0:-1], histogram)
plt.title("Graylevel histogram")
plt.xlabel("gray value")
plt.ylabel("pixel count")
plt.xlim(-40, 0)
```

```python
image_array = image_array[~np.isnan(image_array)]
```

```python
image_array = image_array[~np.isnan(image_array)]
t = ski.filters.threshold_otsu(image_array)
print(t)
```

```python
plt.hist(image_array, bins=200, edgecolor='blue')
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
